```python
import pandas as pd
```

```python
may_19 = pd.read_csv("data/JC-201905-citibike-tripdata.csv")
june_19 = pd.read_csv("data/JC-201906-citibike-tripdata.csv")
july_19 = pd.read_csv("data/JC-201907-citibike-tripdata.csv")
aug_19 = pd.read_csv("data/JC-201908-citibike-tripdata.csv")
sept_19 = pd.read_csv("data/JC-201909-citibike-tripdata.csv")

may_20 = pd.read_csv("data/JC-202005-citibike-tripdata.csv")
june_20 = pd.read_csv("data/JC-202006-citibike-tripdata.csv")
july_20 = pd.read_csv("data/JC-202007-citibike-tripdata.csv")
aug_20 = pd.read_csv("data/JC-202008-citibike-tripdata.csv")
sept_20 = pd.read_csv("data/JC-202009-citibike-tripdata.csv")
sept_20.head()
```

```python
combined_df = pd.concat([may_19,june_19,july_19,aug_19,sept_19,may_20,june_20,july_20,aug_20,sept_20])
combined_df['date'] = pd.to_datetime(combined_df['starttime']).dt.date
combined_df['month'] = pd.to_datetime(combined_df['starttime']).dt.month
combined_df['year'] = pd.to_datetime(combined_df['starttime']).dt.year
combined_df.head(50)

combined_df.to_csv(r'data/full_data.csv', index = False)
```

# Trip Duration grouped by Station start and end Month and year

```python
duration_df = combined_df
duration_df = duration_df.drop(columns=['starttime', 'stoptime', 'bikeid', 'gender'])
trip_duration = duration_df.groupby(['start station id','end station id', 'month', 'year'],).agg({'start station id':['size'], 'tripduration': ['mean'], })
new_duration = trip_duration.reset_index() 
new_duration.head(50)
```

```python
trip_duration.to_csv(r'data/trip_duration.csv', index = False)
```
